```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
N = 1 # number of samples
T = 10 # sequence length
D = 3 # Number of input features
X = np.random.randn(N, T, D)
```

```python
K = 2 # number of output units
M = 5 # number of hidden units

i = tf.keras.layers.Input(shape=(T, D))
x = tf.keras.layers.SimpleRNN(M) (i)
x = tf.keras.layers.Dense(K) (x)

model = tf.keras.models.Model(i, x)
```

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
model.summary()
```

```python
for w in model.layers[1].get_weights():
    print(w.shape)
```

```python
Wx, Wh, bh = model.layers[1].get_weights()
Wo, bo = model.layers[2].get_weights()
```

```python
# trying to replicate the model results
h_last = np.zeros(M) # initial hidden state
x = X[0]
y_preds = []

for t in range(T):
    h = np.tanh(x[t].dot(Wx) + h_last.dot(Wh) + bh)
    y = h.dot(Wo) + bo # This value is only relevent on the last iteration
    y_preds.append(y)

    # re-assign h_last
    h_last = h

print(y_preds[-1])
```
